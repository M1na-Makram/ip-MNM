# 📊 Heart Disease Exploratory Data Analysis (EDA)
This notebook performs a statistical summary and visual analysis of the raw clinical data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os

# Ensure we are in the correct path to load data
data_path = '../data/raw_data.csv'
df = pd.read_csv(data_path)
df.head()
```

### 1. Statistical Summary

```python
print("Dataset Shape:", df.shape)
df.describe().T
```

### 2. Correlation Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='RdBu_r', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

### 3. Feature Distributions (Histograms & Boxplots)

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Age Distribution
sns.histplot(data=df, x='age', hue='target', kde=True, ax=axes[0])
axes[0].set_title('Age Distribution by Target')

# Cholesterol Boxplot
sns.boxplot(data=df, x='target', y='chol', ax=axes[1])
axes[1].set_title('Cholesterol vs Target Outliers')

plt.show()
```
